```python
# Paketləri import edirik
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, color
import cv2

print("Bütün paketlər uğurla yükləndi!")

# Test üçün nümunə şəkil göstərək
# skimage-dən sample şəkil
image = color.rgb2gray(data.astronaut())  # RGB şəkili grayscale-ə çevir
plt.imshow(image, cmap='gray')
plt.title("Test Image: Astronaut")
plt.axis('off')
plt.show()

# OpenCV ilə sadə əməliyyat
image_cv = cv2.cvtColor(data.astronaut(), cv2.COLOR_RGB2BGR)
print("OpenCV ilə şəkil çevrilməsi uğurla tamamlandı!")
```

```python
# Rəngli şəkil nümunəsi
image_color = data.astronaut()  # RGB şəkil
plt.imshow(image_color)          # Rəngli göstər
plt.title("Rəngli Test Şəkil: Astronaut")
plt.axis('off')
plt.show()

# OpenCV ilə BGR formatına çevirmək
image_bgr = cv2.cvtColor(image_color, cv2.COLOR_RGB2BGR)
print("OpenCV ilə RGB → BGR çevrilməsi uğurla tamamlandı!")
```
